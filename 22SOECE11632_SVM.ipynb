```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris
```

```python
iris["Species"].unique()
```

```python
v_nv_fn=lambda x: 0 if x=='versicolor' else 1
iris["v_nv"] =iris["Species"].apply(v_nv_fn)
iris
```

```python
import seaborn as sns
sns.pairplot(iris, x_vars="Petal.Length", y_vars="Petal.Width", hue="v_nv", height=5)
```

```python
#Support vector Classification
from sklearn.svm import SVC
#setting predicters and target
X=iris[["Petal.Length","Petal.Width"]]
y=iris["v_nv"]
#model Building
model=SVC()
model.fit(X,y)
model.score(X,y)
```

```python
pip install mlxtend
```

```python
from mlxtend.plotting import plot_decision_regions
features = np.array(X)
target = np.array(y)
plot_decision_regions(features, target, clf=model)
plt.xlabel("Petal.Length")
plt.ylabel("Petal.Width")
plt.title('Decision Boundary of SVM on iris data')
```

```python
#Encoding 
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
y=iris['Species']
iris['Species'] = label.fit_transform(y)
```

```python
iris['Species'].unique()
```

```python


# Assuming iris is already loaded
X = iris[["Petal.Length", "Petal.Width"]]
y = iris['Species']

model = SVC()
model.fit(X, y)
print(model.score(X, y))
```

```python
from mlxtend.plotting import plot_decision_regions
features = np.array(X)
terget = np.array(y)
plot_decision_regions(features, terget, clf=model)
plt.xlabel("Petal.Length")
plt.ylabel("Petal.Width")
plt.title('Decision Boundary of SVM of SVM on dataframe')
```
