```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load Dataset
df = pd.read_csv('computers.csv')

# Prepare Input (X) and Output (y)
X = df[['Units']]  # Predictor (2D array)
y = df['Minutes']   # Target (1D array)

# Create and Train Linear Regression Model
model = LinearRegression()  # Initialize model
model.fit(X, y)             # Train model

# Get Model Parameters
intercept = model.intercept_
coefficient = model.coef_[0]

# Add Predictions to DataFrame for Visualization
df['Predicted_Minutes'] = model.predict(X)

# Visualize Model
plt.figure(figsize=(10, 6))
plt.scatter(df['Units'], df['Minutes'], color='blue', label='Actual Data')

plt.plot(df['Units'], df['Predicted_Minutes'], color='red', linewidth=2, 
         label=f'Regression Line: Minutes = {intercept:.2f} + {coefficient:.2f}*Units')
plt.xlabel('Units Repaired')
plt.ylabel('Time Taken (Minutes)')
plt.title('Linear Regression Model: Units vs. Repair Time')
plt.legend()
plt.grid(True)
plt.show()

# Compute Coefficient of Determination (Accuracy)
r_squared = model.score(X, y)
print(f"\nModel Parameters:")
print(f"- Intercept: {intercept:.2f}")
print(f"- Coefficient: {coefficient:.2f}")
print(f"\nModel Accuracy (R² Score): {r_squared:.4f}")
```
